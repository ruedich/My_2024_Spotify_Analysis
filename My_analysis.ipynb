# Analysis of my 2024 data

### Importing libraries and data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import pandas as pd

%matplotlib inline 
```

```python
path_data_1 = "my_spotify_data/Spotify Extended Streaming History/Streaming_History_Audio_2024_4.json"
path_data_2 = "my_spotify_data/Spotify Extended Streaming History/Streaming_History_Audio_2022-2024_3.json"
path_data_3 = "my_spotify_data/Spotify Extended Streaming History/Streaming_History_Video_2018-2024.json"

with open(path_data_1, 'r', encoding="utf8") as file:
    data_24_audio_1 = json.load(file)

with open(path_data_2, 'r', encoding="utf8") as file:
    data_24_audio_2 = json.load(file)

with open(path_data_3, 'r', encoding="utf8") as file:
    data_24_video = json.load(file)
```

```python
data_24_audio_1 = pd.DataFrame.from_dict(data_24_audio_1)
data_24_audio_2 = pd.DataFrame.from_dict(data_24_audio_2)

# Convert timestamp to datetime
data_24_audio_1['ts'] = pd.to_datetime(data_24_audio_1['ts'])
data_24_audio_2['ts'] = pd.to_datetime(data_24_audio_2['ts'])

# Filter for entries in 2024
data_24_audio_2 = data_24_audio_2[data_24_audio_2['ts'] >= '2024-01-01']
```

```python
data_24_audio_1.head()
```

```python
data_24_audio_2.head()
```

```python
data_24_video = pd.DataFrame.from_dict(data_24_video)

# Convert timestamp to datetime
data_24_video['ts'] = pd.to_datetime(data_24_video['ts'])

# Filter for entries before December
data_24_video = data_24_video[data_24_video['ts'] >= '2024-01-01']

data_24_video.head()
```

```python
data_24 = pd.concat([data_24_audio_1, data_24_audio_2, data_24_video], ignore_index=True)

# Sort by the 'ts' column
data_24 = data_24.sort_values(by='ts')

# Reseting the index
data_24 = data_24.reset_index(drop=True)

# Filtering before december
data_24 = data_24[data_24['ts'] < '2024-12-01']

data_24
```

## TOP ARTISTS

### 1. TIMES PLAYED (Ocurrences)

```python
top_artists = data_24['master_metadata_album_artist_name'].value_counts().head(10)
print(top_artists)
```

```python
sns.barplot(x=top_artists.values, y=top_artists.index, palette='mako')
plt.title('Times played')
plt.xlabel('Artist')
plt.ylabel('Play Count')
plt.show()
```

### 2. HOURS PLAYED

```python
# Ensure necessary columns are present
if 'master_metadata_album_artist_name' in data_24.columns and 'ms_played' in data_24.columns:
    
    # Group by artist and sum the playtime
    artist_playtime = data_24.groupby('master_metadata_album_artist_name')['ms_played'].sum()
    
    # Convert playtime from milliseconds to hours
    artist_playtime_hours = artist_playtime / (1000 * 60 * 60)
    
    # Sort in descending order
    most_played_artists = artist_playtime_hours.sort_values(ascending=False).head(10)
    
    # Display the top 10 most-played artists
    print(most_played_artists)
    
    # Plot the results
    plt.figure(figsize=(10, 6))
    sns.barplot(x=most_played_artists.values, y=most_played_artists.index, palette='mako')
    plt.title('Top 10 Most Played Artists (by Playtime)')
    plt.xlabel('Total Playtime (Hours)')
    plt.ylabel('Artist')
    plt.show()
else:
    print("Required columns ('master_metadata_album_artist_name' and 'ms_played') are not present in the data.")
```


## STREAMING EVOLUTION

### 1. Daily playtime evolution

```python
data_24['date'] = data_24['ts'].dt.date
daily_playtime = data_24.groupby('date')['ms_played'].sum() / 1000 / 60  # Convert ms to minutes

daily_playtime.plot(figsize=(12, 6), color='green')
plt.title('Daily Playtime (Minutes)')
plt.xlabel('Date')
plt.ylabel('Minutes Played')
plt.show()
```

### 2. Most streamed day

```python
print(f"Most streamed day: {daily_playtime.idxmax()}, with {daily_playtime[daily_playtime.idxmax()]:.2f} mins played")
```

### 3. Monthly playtime progression

```python
data_24_month = data_24

# Ensure the 'ts' column is in datetime format
data_24_month['ts'] = pd.to_datetime(data_24_month['ts'])

# Extract month information
data_24_month['month'] = data_24_month['ts'].dt.month  # Extracts month as an integer (1-12)
```

```python
# Total Playtime by Month
monthly_playtime = data_24_month.groupby('month')['ms_played'].sum() / (1000 * 60 * 60)  # Convert to hours

plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_playtime.index, y=monthly_playtime.values, palette='viridis')
sns.lineplot(x=monthly_playtime.index-1, y=monthly_playtime.values, marker='o', color='blue')
plt.title('Total Playtime by Month')
plt.xlabel('Month')
plt.ylabel('Total Playtime (Hours)')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

## TOP ARTISTS

### 1. TIMES PLAYED (Ocurrences)

```python
# Group by month and artist, count occurrences
monthly_artist_counts = data_24_month.groupby(['month', 'master_metadata_album_artist_name']).size()

# Extract top artist for each month
top_artist_per_month = monthly_artist_counts.groupby('month').idxmax()

top_artists_per_month_df = pd.DataFrame(top_artist_per_month.tolist(),
                                      columns=['month', 'master_metadata_album_artist_name'])

# Add a column for the number of times these artists were played
top_artists_per_month_df['play_count'] = top_artists_per_month_df.apply(
    lambda row: monthly_artist_counts[row['month'], row['master_metadata_album_artist_name']],
    axis=1)

top_artists_per_month_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y=top_artists_per_month_df['master_metadata_album_artist_name'], x=top_artists_per_month_df['play_count'], palette='mako')
plt.title('Top artist per month')
plt.ylabel('Artist')
plt.xlabel('Play Count')
plt.show()
```

### 2. HOURS PLAYED

```python
# Group by month and artist, count time played
monthly_artist_counts = data_24_month.groupby(
    ['month','master_metadata_album_artist_name'])['ms_played'].sum()

# Extract top artist for each month
top_artist_mins_per_month = monthly_artist_counts.groupby('month').idxmax()
top_artists_mins_per_month_df = pd.DataFrame(top_artist_per_month.tolist(),
                                      columns=['month', 'master_metadata_album_artist_name'])

# Add a column for the number of times these artists were played
top_artists_mins_per_month_df['ms_played'] = top_artists_per_month_df.apply(
    lambda row: monthly_artist_counts[row['month'], row['master_metadata_album_artist_name']],
    axis=1) / (1000 * 60 * 60) # from milliseconds to hours

top_artists_mins_per_month_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists_mins_per_month_df['ms_played'],
            y=top_artists_mins_per_month_df['master_metadata_album_artist_name'], palette='viridis')
plt.title('Top artist per month (by Playtime)')
plt.xlabel('Total Playtime (Hours)')
plt.ylabel('Artist')
plt.show()
```
